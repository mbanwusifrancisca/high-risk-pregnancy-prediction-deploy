# 03 — Data Preprocessing

**Objective:** Prepare the synthetic maternal health dataset for model training through encoding, scaling, splitting, and class imbalance handling.

## Steps
1. Load the synthetic dataset
2. Encode categorical variables
3. Train/Validation/Test split (70/15/15)
4. Feature scaling (StandardScaler)
5. Handle class imbalance (SMOTE)
6. Save processed datasets

```python
import sys
sys.path.insert(0, '..')

import numpy as np
import pandas as pd
from src.data.preprocess import (
    load_dataset, encode_categorical, get_feature_target_split,
    split_data, scale_features, handle_imbalance,
    save_processed_data, save_scaler
)

%matplotlib inline
```

```python
# Load dataset
df = load_dataset('../data/synthetic/maternal_health_synthetic.csv')
df.head()
```

```python
# Encode categorical features
df_encoded, label_encoder = encode_categorical(df)
df_encoded.head()
```

```python
# Split features and target
X, y = get_feature_target_split(df_encoded, target_col='risk_label', exclude_cols=['patient_id'])
print(f'Features shape: {X.shape}')
print(f'Target shape: {y.shape}')
print(f'Feature columns: {list(X.columns)}')
```

```python
# Train/Val/Test split
X_train, X_val, X_test, y_train, y_val, y_test = split_data(X, y)
```

```python
# Scale features
X_train_scaled, X_val_scaled, X_test_scaled, scaler = scale_features(X_train, X_val, X_test)
```

```python
# Handle class imbalance with SMOTE
X_train_resampled, y_train_resampled = handle_imbalance(X_train_scaled, y_train, method='smote')
```

```python
# Save processed data
save_processed_data(X_train_scaled, X_val_scaled, X_test_scaled, y_train, y_val, y_test, '../data/processed/')
save_scaler(scaler, '../data/processed/scaler.joblib')
```
