# 01 — Synthetic Data Generation

**Objective:** Generate a synthetic maternal health dataset that reflects realistic distributions and correlations of maternal clinical indicators for Nigerian women.

## Steps
1. Load feature definitions and risk criteria from config
2. Generate synthetic records using statistical distributions
3. Apply risk labeling logic
4. Validate distributions against published reference ranges
5. Save the dataset

```python
import sys
sys.path.insert(0, '..')

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from src.data.generate_synthetic_data import generate_dataset, save_dataset

sns.set_style('whitegrid')
%matplotlib inline
```

```python
# Generate the synthetic dataset
df = generate_dataset(n_samples=5000, random_seed=42)
print(f'Dataset shape: {df.shape}')
df.head()
```

```python
# Summary statistics
df.describe()
```

```python
# Risk label distribution
print('Risk Label Distribution:')
print(df['risk_label'].value_counts())
print(f'\nHigh-risk prevalence: {df["risk_label"].mean():.2%}')

fig, ax = plt.subplots(figsize=(6, 4))
df['risk_label'].value_counts().plot(kind='bar', color=['#2ecc71', '#e74c3c'], ax=ax)
ax.set_xticklabels(['Low Risk (0)', 'High Risk (1)'], rotation=0)
ax.set_title('Risk Label Distribution')
ax.set_ylabel('Count')
plt.tight_layout()
plt.show()
```

```python
# Validate key distributions
continuous_features = ['maternal_age', 'systolic_bp', 'diastolic_bp', 'hemoglobin', 
                       'bmi', 'fasting_blood_glucose', 'weight_gain']

fig, axes = plt.subplots(3, 3, figsize=(14, 10))
axes = axes.flatten()

for i, col in enumerate(continuous_features):
    df[col].hist(bins=40, ax=axes[i], color='steelblue', alpha=0.7, edgecolor='white')
    axes[i].set_title(col)
    axes[i].axvline(df[col].mean(), color='red', linestyle='--', label=f'mean={df[col].mean():.1f}')
    axes[i].legend(fontsize=8)

for j in range(len(continuous_features), len(axes)):
    axes[j].set_visible(False)

plt.suptitle('Distribution Validation — Continuous Features', fontsize=14)
plt.tight_layout()
plt.show()
```

```python
# Save the dataset
save_dataset(df, '../data/synthetic/maternal_health_synthetic.csv')
```
